```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
from keras.layers import Dense
from keras.models import Sequential
from keras.layers import Dropout
from keras.callbacks import EarlyStopping
```

```python
chk=pd.read_csv('/content/ChickWeight.csv')
chk
```

```python
chk.describe()
```

```python
chk.isnull().sum()
```

```python
X=chk.iloc[:,:3]
X
```

```python
Y=chk.iloc[:,3:]
Y
```

```python
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
Y = np_utils.to_categorical(encoded_Y)
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 20, random_state = 1)
```

```python
X_train
```

```python
callback = EarlyStopping(monitor='loss', patience=3)
model = Sequential()
model.add(Dropout(0, input_shape=(3, )))
model.add(Dense(4, input_dim = 3, activation = 'relu',kernel_regularizer='l2'))
model.add(Dense(10, activation = 'relu',kernel_regularizer='l2'))
model.add(Dense(10, activation = 'relu',kernel_regularizer='l2'))
model.add(Dense(10, activation = 'relu',kernel_regularizer='l2'))
model.add(Dense(4, activation = 'softmax',kernel_regularizer='l2'))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
history = model.fit(X_train, Y_train, epochs = 300, batch_size = 59,callbacks=[callback]) 
```

```python
loss, accuracy = model.evaluate(X_train, Y_train)
```

```python
print("Accuracy: %.2f%% Loss: %.2f%%" % (accuracy*100, loss*100))
```
